### What is Clustering

Clustering's function is to divide data points into a number of groups so that data points in the same groups are more similar to other data points in the same group than those in other groups. In other words, clustering is the computational approach to uncover structure in a dataset. The aim is to segregate groups with similar traits and assign them into clusters.



### What is a Cluster?

A cluster is most commonly described as groups of points that are similar or close to each other, however the notion of a “cluster” is not well defined beyond the definition above
  
  
  
 Clusters may be at times subjective. What the concept of boundary around  a set of similar points is subjective. Once we see a cluster we "kind of" know what it is. The objective of cluster analysis is to recognize the similarities or differences between groups of objects and describe them in an algebraic or graphic form in order to have a better understanding of a given domain.

```python
## Here we define how to calculate the Jaccard distance

def compute_Jaccard_distnace(vec_1, vec_2 ):
    
    # All the symptoms shared by both
    intersection = sum((vec_1 == vec_2) & (vec_1 == 1))
    # All the symptoms that either one has
    union = sum((vec_1 == 1) | (vec_2 == 1))
    return 1 -  intersection / union
    
```

###  Devising you Own Distance

When defining your own distance metric, you must follow certain properties that any function should have:

 $$
 d(x, y) \ge 0 \\
 d(x,y)=0 \text{if and only if} d(x, y) = d(y, x) \\
 d(x, y) + d(y, z) \ge d(x, z)
 $$ 
 
 For technical reasons, the symmetry property is usually highly desirable. You can always construct a symmetric distance function from an asymmetric one
 
 $$
     dS(x,y)= \frac{d(x,y)+d(y,x)}{2}
 $$
 
 The following link gives us an excellent example how the most common distance metrics are implemented.  With such knowledge, it becomes easier to do the aforementioned.
 
 https://scikit-learn.org/stable/modules/clustering.html

### Smooth Function are Better than Non-smooth Ones

Given a basic distance function:
 
 $$
 d(x,y) = \sum{d_i(x_i, y_i)}~~\mbox{where,}\\
 d_i(x_i, y_i)= \left\{
     \begin{array}{ll}
         1 & \mbox{if } x=y \\
         0 & \mbox{otherwise}
     \end{array}
 \right.
 $$
 
 We can immediately conclude it is not "smooth."  This is because the difference between points [1,2] and [2,3] are the same as [1,2] and [100,200].  Intrinsically, this should not be the case.  However, with the definition we provided the distance will equate to 0 as x does not equal y.

# Silhouette Coefficient

## What's so useful about it?

When you have a set of distinct clusters, visually (as humans), we may be able to determine whether or not each cluster is a distinct cluster. More importantly of course, we can also intuitively determine whether or not the clusters are _good_ sets of clusters.

### Consider the clusters


These clusters make sense, but would it be better for some of the blue points to be considered as part of the red cluster? What about an entire merge of the two clusters?
When it's left to opinion, it may be pretty difficult to say for sure and answers may differ from person to person. The **silhouette coefficient** thereby serves as a means to put an objective value towards the quality of each point belonging to their cluster and consequently, the quality of the cluster itself.

### Calculation
Let $a_{i}$ be the average distance that point $i$ has from all other points in the same cluster (cohesion - lower is better),
and $b_{i}$ be the smallest distance between $i$ and a point not in the same cluster as $i$ (separation - higher is better)

The silhouette coefficient of a point $i$ is therefore calculated as:
$$S_i = \frac{b_i - a_i}{max(a_i, b_i)}$$

When subtracting the average distance of all points in the same cluster as a provided point from the smallest distance to another cluster, we can get the understanding that the larger the number, the better off we are. Dividing the maximum of the two serves as a means to normalize the result, so its comparable with other points within the same cluster.

As a result, values range from -1 to 1; positive values correspond to high cohesion and/or high separation, whereas negative values corresponds to low cohesion and/or low separation.


## Understanding the values of a silhouette

### Consider the clusters and silhouette plot



The fifth cluster (orange) has a negative silhouette when the point in question has a larger average distance to points within its own cluster, but the shortest distance to another cluster is less than that average distance. The points those values belong to are likely the few close to cluster 1 (blue). Likewise, the points that are closest to its centroid will have the larger silhouette values.

This does provide the notion that the closer a point is to 1, the more likely it's true that the point belongs within that cluster. Likewise, the closer it is to -1, the more likely it belongs to a different cluster.


### Optimal number of clusters

We know now that at the per-point level, we can figure out how well a given point belongs to their own cluster using its silhouette. By graphing out the silhouette of every single point of every cluster, we can get a pretty good understanding of how well a provided number of clusters fits with each point using the mean value of all silhouettes. This means we can actually run a silhouette scoring for any K-cluster to find an optimal number of clusters.

An example of this below provided by: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html

```python
from __future__ import print_function

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

print(__doc__)

# Generating the sample data from make_blobs
# This particular setting has one distinct cluster and 3 clusters placed close
# together.
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)  # For reproducibility

range_n_clusters = [2, 3, 4, 5, 6]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```
