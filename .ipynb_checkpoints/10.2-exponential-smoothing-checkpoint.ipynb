### Random Walk and White Noise

White noise is a time series in which each point has no correlation with the rest of the points in the series. White noise has a constant mean and variance. Gaussian white noise ($ e \sim \mathcal{N}(\mu, \sigma) $ are points taken from a Gaussian distribution. White noise cannot be predicted but the mean of the noise can be predicted. White noise exists in a time series forecast model as the error term. In a random walk, the change in value from point to point is white noise. Random walks are a time series model where each point was generated by adding white noise to the previous point and is defined with the formula $ v_t = v_{t-1} + \epsilon_t, $ where $\epsilon_t$ is the white noise.


```python
plt.figure(figsize=(20,6))

v = np.zeros(1000)
w = np.zeros(1000)
for t in range(1, len(v)):
    w[t] = np.random.normal(0,1)
    v[t] = v[t-1] + w[t]
    
txt="Plot of white noise and random walk"
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=12)
plt.plot(np.arange(1000),v)  
plt.plot(np.arange(1000),w)   
```

## Forecasting
The naive method is that variable at time $t$ can be predicted using time $t-1$ but this can lead to an inaccurate prediction in the case that your prediction is outside of the trend. This is why the sliding windows method can be used. The sliding windows method is using the previous points to predict the following point, but to do so the white noise needs to be abstracted. This can be done by making $v_t$ the mean of the previous points ($
v_t = \frac{1}{n}\sum_{i=0}^{n-1}v_{t-i}
$) This is called smoothing the data, which means taking the trend out so that the seasonality can be analyzed.


### Predicting Random Walks -- Cont'd

* The best guess for $v_t$ is $v_{t-1}$

$$
\hat{v}_{t} = v_{t-1}
$$

* This is the "state of the art" for handling data from random walks 
 * Maybe impossible to beat, even with much more sophisticated  solutions
  * Also, called the nave method or random walk forecasts
  
  



```python
plt.figure(figsize=(20,6))
v2 = np.zeros(100)
for t in range(1, len(v2)):
    v2[t] = v2[t-1] + np.random.normal(0,1)

txt="Naive method"
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=12)
plt.plot(np.arange(100),v2)  
plt.plot(np.arange(1,101),v2)  
```

### Smoothing Using a Running or Moving Average

* Variations exist around how to determine 
  * The window strategy
    
Window strategies, or otherwise known as exponential window functions, are used to assign exponentially decreasing weights over time. By applying the exponentially decreasing weights, data can be smoothened by removing high frequency noise that would otherwise disproportionately affect the data. With this procedure, drastic effects of outliers diminish over time.
     
* Example 1:     
For some window of size 2k+1, we compute $s_i$ as: 

$$
s_i = \frac{1}{2k+1} \sum_{j=-k}^{k}x_{i+j}
$$

* Ex. for $k=3$, the window size is $2k+1=7$, 

$$
s_i = \frac{1}{7} \sum_{j=-3}^{3}x_{i+j},
$$

and computing the smoothed score $s_{11}$ at position $x_{11}$ amount to averaging the score for positions:

$$
x_{8}, x_{9}, x_{10}, x_{11}, x_{12}, x_{13}, x_{14},
$$

which is also known as the range of positions from j (3 in this example) units left of $x_{11}$ ($x_{8}$ in this example) to j units right of $x_{11}$ ($x_{14}$ in this example) inclusive

* Example 2:     
For some window of size k, we compute $s_i$ as: 

$$
s_i = \frac{1}{k} \sum_{j=0}^{k}x_{i+j}
$$

* Ex. for $k=4$,  

$$
s_i = \frac{1}{4} \sum_{j=0}^{4}x_{i+j},
$$

and computing the smoothed score $s_{11}$ at position $x_{11}$ amount to averaging the score for positions:

$$
x_{8}, x_{9}, x_{10}, x_{11},
$$

which is also known as the range of positions from j-1 (4-1=3 in this example) units left of $x_{11}$ ($x_{8}$ in this example) to $x_{11}$ inclusive

### The effect of the Window Size

* The rolling window smooths the data, but the curve is still bumpy around large spikes
  * As seen in the previous plot, the outliers were too great to dampen.
  
* Sudden jumps in the MA value when a relatively large observation (a spike) or a a relatively small observation (dip) comes into or falls out of the moving average window.
  * Spikes will greatly increase the MA while dips will greatly decrease the MA unless otherwise dealt with.

* We can further smooth the curve by choosing a larger window size
  * More values to dampen the effect of larger values
     * More lower values to reduce spikes, and more higher values to reduce dips
     * These values are obviously adjacent to the spike/dip, in order to have the greatest impact under exponential smoothing.
     
  * Results in under emphasizing real trends as well
     * This does not take into account false positives and false negatives that may actually be accurate data instead of just random noise
     * False positive - error in data in which a test result improperly indicates presence of a condition when it really is not present
     * False negative - error in which a test result improperly indicates no presence of a condition, when it really is present
     * In this case, a false positive would mean identifying an increase as a spike and a decrease as a dip when they are not, and a false negative would mean identifying dips and spikes as data.

```python
plt.figure(figsize=(18,6))
plt.xlim(0, len(random_ts))
plt.plot(random_ts)
plt.plot(random_ts['val'].rolling(window=10, center=True).mean())
```

In the plot above, the original data set is blue, while the smoothened data is in orange. Note how the original data set is the same one as the previous plot. However, since the window size here (10) is 2.5 times wider than the previous one (4), the smoothened data is less bumpy and less sharp.

### Handling Outliers


* Increasing the window to compensate for few spikes can lead to underfitting the data

  * Overly "flatten" the curve
    * It may just result in a constant linear line that is too simple to explain variance
  * Tricky (perhaps impossible) to find a perfect fit
    * Like the line of best fit, it may fit for current data, but not so much for newer data
* Unduly biases toward sharp, even legitimate increases or decreases
  * Causes a lag in the signal
     * Ignoring false positives and false negatives delays new trends, causing data to be less accurate

* Rather than increasing the window size, we could, for instance, weigh the values in the window differentially
     * For example, weighing values by their amount of occurances in the window
* The point close to window boundaries contribute less than the point in the center 
     * As the center of the window contain the most occurances while boundary points contain the least
     * Exactly like some sort of distribution...

### Using a Gaussian for Computing the Contributions 


$ s_i = \sum_{j=-k}^{k}w_{i+j}x_{i+j} $ 

where weight $w_i$ for a point $x_i$ is determined using a Gaussian, such as:
$$
f(\mu_x, \sigma_x) = \frac{1}{\sqrt{2\pi\sigma_x^2}}exp(-\frac{1}{2}(\frac{\mu_x}{\sigma_x})^2)
$$

By using the Gaussian, the contribution of each point in the window can be computed.

For more information about normal/Gaussian distribution:
http://128.171.10.138:8080/url/128.171.10.138:4000/assets/ipynb/week_4/2_normal_distributions.ipynb
https://en.wikipedia.org/wiki/Gaussian_function

### Challenges of Using a Gaussian for Computing the Contributions 


* The range of values in each window can vary substantially
  * And this can change the moving average
  * Every window has its own different outliers/spikes/dips/false positives and negatives
* For each window, we need to dynamically scale $\sigma$ so that the likelihoods are not too small
  * If the likelihoods are too small, they may not be included in calculating the Gaussian
  * Without including them, those points will not be used to compute the contributions

### Challenges of Using a Gaussian for Computing the Contributions  - Cont'd

* Easy to implement but not ideal:

  * The weighted MA is "painful" to evaluate 
    * For each window, we need to determine a specific $\sigma$ and compute the likelihoods for all the points in the window
    * As more data is added, the windows, their means, and likelihoods all have to be recomputed
  * We lose data at the beginning and the end of data due to window constraints
    * There is no data before the start and after the end of the data
    * An infinite sample size is impossible
* No easy way to incorporate information about the trend or seasonality
    * Every outlier has to be considered


### Exponential Smoothing

* Exponential smoothing mitigates the shortcomings of the rolling window and weighted rolling window. This is done by using all past observations as oppose to a subset of the data. The contributions (weights) associated with each point, decays exponentially as the observations get older. This simple idea layes the foundation for some of the most successful forecasting methods. 



### Forms of Exponential Smoothing




| Type of Smoothing            | Data type                     |
|:------------------------------|:-------------------------------|
| Single Exponential Smoothing | Neither trend nor seasonality |
| Double Exponential Smoothing | Trend but no seasonality      |
| Triple Exponential Smoothing | Trend and seasonality         |


### Single Exponential Smoothing

Computes smoothed data $s_t$ at time $t$ using the observed value $v_t$ and smoothed value from previous time step ($v_{t-1}$)


$$
s_t = \alpha~v_t + (1-\alpha)~s_{t-1} ~~~~~~\text{where}~~~~~~ 0 \le \alpha \le 1
$$



$s_t$: smoothed value at timestep $t$  
$v_t$: actual unsmoothed value at time step $t$  
$s_{t-1}$: smoothed value at time step $t-1$



### Single Exponential Smoothing: Remarks
  
Note that at:

* $\alpha =0, s_t = (1-\alpha)~s_{t-1}$ we are only retaining the past
* $\alpha =1, s_t = \alpha v_t$ we are only retaining current values

How is this exponential? 
    
 -  All the previous values contribute to smoothed value of $x_i$ but their contribution as increasingly small

It turns out that:

$$
\begin{align}
s_t &= \alpha~v_t + (1-\alpha)s_{t-1} \\
    &=  \alpha~v_t + (1-\alpha) (\alpha~v_{t-1} + (1-\alpha)s_{t-2})\\
    &=  \alpha~v_t + \alpha(1-\alpha)~v_{t-1} + (1-\alpha)^2s_{t-2}\\    
    &=  \alpha[v_t + (1-\alpha)v_{t-1} + (1-\alpha)^2v_{t-2}+ \ldots + (1-\alpha)^{t-1}v_{1}]+(1-\alpha)^t v_{0}\\
\end{align}
$$


### Single Exponential Smoothing: Remarks -- Cont'd
  

$$
\begin{align}
s_t  &=  \alpha[v_t + (1-\alpha)v_{t-1} + (1-\alpha)^2v_{t-2}+ \ldots + (1-\alpha)^{t-1}v_{1}]+(1-\alpha)^t v_{0}\\
\end{align}
$$

* The contribution of past values is controlled by the decay of the parameter $\alpha$
 * The speed of this decay is slower when $\alpha$ is closer to 0 and faster whne $\alpha$ is cloer to 1. 
  * This geometric progression is the discrete version of an exponential function,
    * This is where the name for this smoothing method originated




```python
plt.figure(figsize=(18,6))
plt.plot(random_ts)
plt.plot(random_ts['val'].ewm(alpha=0.5).mean(), color='k', lw=5,)
plt.plot(random_ts['val'].rolling(window=5, center=True).mean(), lw=5)
```

### Double Exponential Smoothing


When we have to consider trend information the single exponential smoothing is not enough. To consider trend information we have to apply double exponential smoothing which applies an exponential filter twice. The parameter $\alpha$ is the data smoothing factor and $\beta$ is the trend smoothing factor. Both $\alpha$ and $\beta$ serve as a measure of how much of the past we want to consider. The more we weight the past, the less we weight the current time point and vice versa. Double Exponential Smoothing takes into account trend, but not seasonality. $s_t$ computes the level, but also takes into account trend. $r_t$ computes the trend.
  

$$
s_t = \alpha v_t + (1-\alpha)(s_{t-1} + r_{t-1})
$$
$$
r_t = \beta(s_{t} - s_{t-1}) + (1-\beta)r_{t-1}
$$




### Double Exponential Smoothing: The Level

$$ 
s_i = \alpha v_t + (1-\alpha)(s_{t-1} + r_{t-1})
$$

We can see that $s_t$ is almost the same as the equation for single exponential smoothing, but here we are adding the trend information at $t-1$. Thus, $s_t$ now accounts for trend information from the previous time point.


### Double Exponential Smoothing: The Seasonality

$$
r_i = \beta(s_{t} - s_{t-1}) + (1-\beta)r_{t-1}
$$

When computing the trend at $t$ we compute the difference $s_{t} - s_{t-1}$ (the slope). Parameter $\beta$ serves as a way to decide how much of the past we consider when computing the value at time $t$. The lower the $\beta$ the more we consider the previous trend and the less we consider the difference in level between time point $t$ and $t-1$ and vice versa. We can compute the next value by adding the current level and trend:

$$
v_{t+1} = s_t + r_t
$$




### Triple Exponential Smoothing

When we want to consider seasonality as well we want to consider previous values that share the same period. One way to think about this is by looking at sine function where the x-axis are months. If we want to smooth for the month of March we will want to consider all previous points from the month of March. So we smooth point at time $t$ by considering the point at time $t-p$ where $p$ takes the value of a period (if we were doing smoothing for March we could set this to 12 so for every point in March we would consider only that month in the previous year).


### Triple Exponential Smoothing -- Cont'd

The new equation for the triple exponential component is:
$$
    s_t = \alpha v_t + (1-\alpha)(s_{t-1} + r_{t-1})
    $$
    $$
    r_t = \beta(s_{t} - s_{t-1}) + (1-\beta)r_{t-1}
    $$
    $$
    p_t = \gamma(v_t - s_t) + (1-\gamma) p_{t-k}
$$


When we want to consider seasonality as well we add an extra equation that behaves in a similar way as the previous two equations. Equation $p_t$ uses parameter $\gamma$ where $ 0 \leq \gamma \leq 1$. The choice of $\gamma$ will also determine how much of the past we consider when computing for the current time point. The difference here is that the past we consider is only the point $k$ time points previous to the current, where k is the period.


### Triple Exponential Smoothing: Remarks


If we want to predict a point that is $h$ timesteps ahead of $t$ the seasonality of the point at $t+h$ is dependent on the point observed exactly one period $k$ previous to $t$. This condition makes use of the symmetry of the seasonality: two points with exactly one period difference are assumed to have the same seasonal value (i.e. the sine function). The modulo is used to ensure that the predictions are based on current or past observations, not from other predictions. We can compute future values with the following:

$$
v_{t+h} = s_t + h~r_t + p_{t-k+1+((h-1) \mod k)} 
$$



### Testing the Explonential Smoothing

* Does the approach pass muster with the CO2 data set we manually investigated

  * The data exhibits clear trend and seasonality

* Here we implement and test our owen version 

  * You will use the statsmodel in the assignmnet 


```python
# function that returns the initial value of trend
def initial_trend(series, slen):
    sum = 0.0
    for i in range(slen):
        sum += float(series[i+slen] - series[i]) / slen
    return sum / slen

# function that return the initial value of seasonality
def initial_seasonal_components(series, slen):
    seasonals = {}
    season_averages = []
    n_seasons = int(len(series)/slen)
    
    # compute season averages
    for j in range(n_seasons):
        season_averages.append(sum(series[slen*j:slen*j+slen])/float(slen))
        
    # compute initial values
    for i in range(slen):
        sum_of_vals_over_avg = 0.0
        for j in range(n_seasons):
            sum_of_vals_over_avg += series[slen*j+i]-season_averages[j]
        seasonals[i] = sum_of_vals_over_avg/n_seasons
    return seasonals

def triple_exponential_smoothing(series, slen, alpha, beta, gamma, n_preds):
    result = []
    seasonals = initial_seasonal_components(series, slen)
    for i in range(len(series)+n_preds):
        if i == 0: # initial values
            smooth = series[0]
            trend = initial_trend(series, slen)
            result.append(series[0])
            continue
        if i >= len(series): # we are forecasting
            m = i - len(series) + 1
            result.append((smooth + m*trend) + seasonals[i%slen])
        else:
            val = series[i]
            last_smooth, smooth = smooth, alpha*(val-seasonals[i%slen]) + (1-alpha)*(smooth+trend)
            trend = beta * (smooth-last_smooth) + (1-beta)*trend
            seasonals[i%slen] = gamma*(val-smooth) + (1-gamma)*seasonals[i%slen]
            result.append(smooth+trend+seasonals[i%slen])
    return result
```

```python
co2_data = [float(x) for x in open("data/carbon_dioxide.txt")]
```

```python
# We use the exponential smoothing to forecast the next 24 months (2 periods)

co2_data = [float(x) for x in open("data/carbon_dioxide.txt")]

plt.figure(figsize=(18,6))


forecast = triple_exponential_smoothing(co2_data, 12, 0.005, 0.2, 0.00001, 24)

plt.plot(forecast, 'r--', lw=4) # the smoothed/predicted data in red

plt.plot(co2_data, alpha=0.5, lw=4) # the true data in blue
```

The example above used for forecasting is fed values for alpha, gamma, and beta. In order to derive these values we can use the curve_fit function from scipy.optimize to get the parameter values. The function takes as inputs the function, our input data, and output data and returns the best parameters that fit to the curve by minimizing the squared error. The documentation can be found here:
https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

Below is an example of checking how well the above parameter values fit. The lower the TSS, the better the fit.

```python

def tss(y, y_hat):
    """
    y:     the true data
    y_hat: the generated data
    
    This function will return the sum of squares error (TSS).
    """
    
    return sum( (np.asarray(y)-np.asarray(y_hat))**2 )

# compute TSS for given parameter values above for the true data and the data generated by the triple exponential
gen_co2 = triple_exponential_smoothing(co2_data, 12, 0.005, 0.2, 0.00001, 0)
tss_init = tss(co2_data, gen_co2)

print(f'TSS for alpha=0.005, beta=0.2, gamma=0.00001: {tss_init}')
```
