### Correlations


Two variables X and Y are correlated if knowledge of one helps better predict the other. 
This will in turn help us predict the outcome of one variable based on the correlation with another. If we know one of the variables we can say more about the other one if they are correlated. The correlation coefficient between $X$ and $Y$ is a numerical value that measures the degree to which $Y$ is a function of $X$ (and vice-versa).


### Examples: 

* Weight is correlated with height (relatively strong positive correlation)
  
  * The number of tourist in a Waikiki is associated with the total sales at ABC stores (reasonably strong positive correlation)
  * IQ and time to required to solve simple logic problems (reasonably strong negative correlation)
  * The number of daily cigarettes smoked and the number of years lived (reasonably strong negative correlation)
  
  * The number of video games played and violent behavior score (very weak positive correlation)
  


### Linear Correlations

Linear correlation means that there is a linear relationship between two variables. This can be written as the form Y = aX + b. In the case where Y and X have a positive or negative correlation of 1, given one variable we can predict the other with certainty that the prediction is correct. If the correlation is closer to 0, the correlation is not irrelevant, but it will be more of a better way of predicting the value of the other variable. An example of this is the correlation between height and weight of a person. If a person is taller than the average person, you know that he is likely to weigh more that a average person. This does not mean that no tall people weigh less than the average weight. The linear relationship gives you the average weight of a person at a certain height, but there is going to be large variances in the weight of all the people at that height. It is still in most cases going to be closer to the actual weight than if you did not have the extra information of the person's height. 

A measure of correlation is the Pearson correlation coefficient. It is on a scale from -1 to 1 where a correlation of 1 means perfect correlation, a correlation of -1 means perfect inverse correlation and the closer the value is to 0 the less correlation there is. If the correlation is 0, one variable gives us absolutely no information about the other, while a correlation of 1 or -1 in theory gives us perfect information about the other.

### Computing The Correlation Coefficient

+ If the correlation is positive: when X's distance from it mean increases, Y's distance from its mean increases as well. 
+ If the correlation is negative: when X's distance from its mean increases, Y's distance from its mean decreases or vice-versa. 
+ In the absence of correlation the contributions of $X_i$ and $Y_i$ cancel out. Therefore, the correlation tends toward 0.

To find the Correlation Coefficient we often use the Pearson Correlation Coefficient. To find it, we take the sum of the product of the distance to the mean of X and Y for all the available data points. We then divide this by a term that makes the number always be between -1 and 1.  

Pearson Correlation Coefficient (https://www.youtube.com/watch?v=ugd4k3dC_8Y).
 

### Computing The Correlation Coefficient - Cont'd

* Pearson Correlation Coefficient

$$
r = \frac{\sum_{i=1}^{n}{(X_i-\overline{X})(Y_i-\overline{Y})     }}{\sqrt{\sum_{i-1}^n{(X_i-\overline{X})}} \sqrt{\sum_{i-1}^n{(Y_i-\overline{Y})}} }
$$




- Pearson correlation measures how **in sync** are two variables $X$ and $Y$

- For two variable to be corrected:

   - If a value from `X` is much bigger than its means, then we want the other one to be equivalently larger than it's mean
   - If a value from `X` is much smaller than its means, then we want the other one to be equivalently smaller than it's mean 
   
- Two variables are not correlated (at least not according to the Pearson Correlation Coefficient) when half of the contributions are above the mean and half are below the mean

A pearson value indicates the extent to which the data plotted on a scatter plot are in a straight line. If the data is in a completely straight line the correlation value would be 1.0 or -1.0 (depending on slop of te line). If the data is completely scattered the value would be 0.0.


### $R^2$ and the Explained Variance

- $R^2$ is  the square of the correlation coefficient

- The interpretation of $R^2$ is different from that of $R$

* Example:

Variance is a technique used in probability theory and statistics that measures how far a set of random numbers differs from the their average values. It is the square of the correlation coefficient $R$ but it's interpretation is not the same. The correlation coefficient measures the linearity between two variables while variance is the measure of what numbers you can expect from the mean.

Some examples that use this technique are fields in statistics such as descriptive statistics, statistical inference, hypothesis testing and much more. An easier example can be something like number of accidents that occur on the H1 versus number of taxis on the highway that same day.

### The regression residuals

* The difference between an observed value and an estimated value

* The error between the observed value and the predicted value (line)

Regression residuals are observations in the data and the regression line. To simply put, this is an observation between the observed value and the estimated value (the line). You can also think of regression residuals as the statistical error between the observed value and the expected value. Because values fluctuate above and below the regression line, the sum and the mean of the residuals is equal to the smallest possible value (0).

As you would expect, it is our desire to minimize the regression residuals and thus enable us to have a stronger correlation. What if we compare the regressional line with the mean line?

### Spearman Correlation

- You can think of it as Pearson correlation but with rank data

- Counts the number of disordered pairs, not how well the data fits a line.

- Better with non-linear relationships and outliers.


### Spearman Correlation

- Let $rank_{x_i}$ be the rank position of $x_i$ in  sorted order
  - $rank_{x_i}$ ranges from 1 to n
  
 $$
 \large
 \rho = 1 - \frac{6 (rank_{x_i} - rank_{y_i})}{n (n^2 -1)}
 $$
 
 - Given two data sets A and B.
 - We the rank them according to values and label X for A and Y for B.
 - Calculate the difference between X and Y,  d=X-Y
 - Square the difference for each value, d^2.
 - Sum all the d^2, which equals $2$
 - n = number of data points, n = 4

 
| A  | B  | X | Y | d=X-Y | d^2 |
|----|----|---|---|-------|-----|
| 34 | 66 | 4 | 3 | 1     | 1   |
| 55 | 88 | 2 | 2 | 0     | 0   |
| 43 | 34 | 3 | 4 | -1    | 1   |
| 75 | 76 | 1 | 1 | 0     | 0   |

$(rank_{x_i} - rank_{y_i}) = 2$

 $$
 \large
 \rho = 1 - \frac{6 (2)}{4 (4^2 -1)}
 $$
 
 $$\rho = 1 - \frac{12}{60}$$
 
 $$\rho = 1 - 0.2$$
 
 $$\rho = 0.8$$
 
 Which means that from the data provided by A and B, for every increase in A there is an increase in B. Our data has one miss step, so 80% of data is correlated.

## Statistical Significance
  
* Data miners are primarily concerned with  whether their observations are interesting

* Statisticians are primarily concerned with whether observations on data are significant

  * As we discussed, something is statistically significant if the odds of it happening by chance are rare
  * Less than 1 in 20 (probability of the event is less than 0.05)


### Interpreting correlation significance

* Statistical significance of a correlation measures how likely it is for data to be corrected just by chance

* Depends on the sample size and the real underlying correlation coefficient

  * Two data points are always correlated ($r=1$) 
  * Much larger dataset are less likely to be correlated just by chance.
  
* Using bootstrapping, we can build our own background distribution of correlation coefficients, which we can use to assess  the statistical significance of any given value

We have to figure out if the linear correlation between the data is strong enough to actually use as a model. The sample size needs to be considered when figuring this out. If there is a correlation that exists between 3 data points, that might just be by chance. If there exists a coorelation between 1million data points, that has less probability happening due to chance. We do this by creating a bootstrapping model  we can use to rule it out chance.

I.e
An experiment was ran that showed adults who watched TV with commercials showing food ate more nuts than the group who wasn't shown food commercials. Was this due to chance? 

We can test this using a boot strapping model that uses the results of the experiment, but randomizes the original label. Then run it a bunch of times. And check if the two groups over all runs show the same result: people shown food commericals ate more nuts.

````python
from scipy import stats

x = np.random.randint(0, 100, 10)
y = np.random.randint(0, 100, 10)
````

````python
corr_ = []

for _ in range(100000):
    positions = np.random.choice(range(10), size=10, replace=True)
    corr_.append(np.abs(stats.pearsonr(x[positions],y[positions])[0]))
````

````python

plt.figure(figsize=(12,8))

weights = np.ones_like(corr_)/float(len(corr_))
_ = plt.hist(corr_,  bins=20, weights=weights, edgecolor='black', linewidth=1.2)
plt.xlabel("Correlation value")
plt.xlabel("Correlation value")
````

````python
def simulate_corr(n = 10, nb_repeats=1000, p=1/20):
    corr_ = []
    for _ in range(nb_repeats):
        x = np.random.randint(0, 100, n)
        y = np.random.randint(0, 100, n)
        corr_.append(np.abs(stats.pearsonr(x,y)[0]))
    return (np.quantile(corr_, (0, 1-p)[1]), max(corr_))
````

````python
 simulate_corr(n = 10, nb_repeats=100000, p=1/20)
````

````python
sample_size = np.arange(10,1010, 10)
max_common_corr = []
max_corr = []
for n in sample_size:
    max_c_r, max_r = simulate_corr(n, nb_repeats=100)
    max_common_corr.append(max_c_r)
    max_corr.append(max_r)
````

````python
plt.figure(figsize=(18,6))
plt.plot(sample_size, max_common_corr)
plt.scatter(sample_size, max_corr, color="red")
````

Below is the same calculation from above but performaing Spearman instead of pearson.

````python
def simulate_corr_spearman(n = 10, nb_repeats=1000, p=1/20):
    corr_ = []
    for _ in range(nb_repeats):
        x = np.random.randint(0, 100, n)
        y = np.random.randint(0, 100, n)
        corr_.append(np.abs(stats.spearmanr(x,y)[0]))
    return (np.quantile(corr_, (0, 1-p)[1]), max(corr_))

sample_size = np.arange(10,1010, 10)
max_common_corr = []
max_corr = []
for n in sample_size:
    max_c_r, max_r = simulate_corr_spearman(n, nb_repeats=100)
    max_common_corr.append(max_c_r)
    max_corr.append(max_r)
    
plt.figure(figsize=(18,6))    
plt.plot(sample_size, max_common_corr)
plt.scatter(sample_size, max_corr, color="red")
````

### Spurious correlations or wrong hypotheses

- However, the above applies if we test a single pair of variable $X$ and $Y$. 
  - Similarly, our reasoning about hypothesis testing applies primarily to single test (two competing hypothesis). 

- The logic changes if we are comparing a much larger set of hypothesis or if we are computing correlation on a large number of random variables.
  
Before we were just testing between two variables. Datasets can contain a large amount of variables and if we just look for correlations between them all without considering statisical significance we can may make a model based on a correlation that has no significance and make incorrect predictions.

- In a large number of trials, even low rates of Type I (false positive) and Type II (false negative) errors can have a significant impact on the results
  - False positives and false negatives are rare
    - How rare will depend on the population's shape
    - In the normal, you can expect a false positive rate of about 0.05
  - A low probability outcome will occur if we repeat the experiment a large number of times


- In a relatively large dataset with 200 variables, one can test:
    - $n (n-1) / 2$ possible correlations
    - $(200 * 199) / 2 = 19900$ possible correlation

- If the event of a correlation arising by chance is very rare; say 0.001
- Then in a dataset with ~20K correlations, we would expect ~ 20 correlations to have arisen by chance
- If the false positive rate is 0.01, which is a very reasonable approximation, then you should expect ~ 200 correlations to have arisen by chance
  - That's a concerning number!

- In the Medicaid dataset, you could compare all medication usage between the east-cost and the west-coast
  - Is there a difference between east coast and west coast of PENICILLIN use
  - Is there a difference between east coast and west coast of ASPIRIN use
  - Is there a difference between east coast and west coast of OXYCODONE use
  - Is there a difference between east coast and west coast of ACETAMINOPHEN use
  .
  . 
  
- In such a large number of tests, Type I and Type II will most likely occur.


#### Mutiple Testing Correction

- What we need is a way to identify and remove those p-values which significance is only due to chance
  - In a way, we need to correct for all the tests we are carrying out


````python
test_pvalues = []
for i in range(10000):
    group_1 = sp.stats.norm(177, 7).rvs(30)
    group_2 = sp.stats.norm(177, 7).rvs(30)
    test_pvalues.append(sp.stats.ttest_ind(group_1, group_2)[1])
````

````python
_ = plt.hist(test_pvalues, bins=20)
print("Out of 10,000 tests, the number of false positives is %s "  % sum(np.array(test_pvalues) < 0.05))
````

````python
test_pvalues = []
for i in range(10000):
    group_1 = sp.stats.norm(177, 7).rvs(30)
    group_2 = sp.stats.norm(174, 4).rvs(30)
    test_pvalues.append(sp.stats.ttest_ind(group_1, group_2)[1])
````

````python
_ = plt.hist(test_pvalues, bins=10, color = 'r')
print("Out of 10,000 tests, the number of false negatives is %s "  % sum(np.array(test_pvalues) > 0.05))
````

# Linear Regression


- A simple method to predict a *quantitative* response $Y$ on the basis of a single predictor variable $X$

- Assume an approximately linear relationship between $X$ and $Y$

- Remember that the equation of a line is:

$$
    Y = \beta_o + \beta_1 ~ X
$$

The linear regression method is used to model the relation between two variables.  Linear regression is a technique which computes a regression line from data.  The regression line is a line which shows a linear relationship between a variable X and a variable Y.  The equation of the regression line, Y=βo+β1*X, where βo is the y-intercept and β1 is the slope, can be used to calculate a value of Y given a value of X.  Linear Regression is useful when there is an relationship between X and Y that is approximately linear.

### Linear Regression

* `linregress` returns the intercept $\beta_0$ and the slope $\beta_1$ of the linear relationship between sales and advertising budget for TV

$$
    sales \approx \beta_o + \beta_1 ~ TV
$$

* Together, $\beta_0$ and $\beta_1$ are known as the model parameters

    sales≈βo+β1*TV

The linregress function can be used to calculate the linear regression values of βo and β1 which are known as the model parameters.  By using this, the number of units sold can be predicted for any given TV advertising budget.

### The Residual Sum of Squares

* We can quantify how well a line fits the data simply by assessing the total error 
* To avoid that error values cancel out, we compute the square of the error
* We call that the residual sum of squares (RSS)

$$
RSS = e_1^2 + e_2^2 + ... + e_n^2
$$

The error from the linear regression model can be used to determine how well the line fits the data.  The error of a data point is it's y-axis distance from the regression line.  Since the error can be either negative or positive, the error on every data point is squared so that all values are positive and then they're added to the total.  The result of these squared error sums is known as the residual sum of squares of RSS. The RSS is what will be used to determine how well the line fits the data.



### Regression coefficients - Minimizing the RSS

* Naturally, we want the model that provides the smallest RSS
  * This not always the case with other modeling techniques
  * A "database" is a "model" which can provide RSS=0, yet, it's a bad model

* The line with the smallest RSS is the best model under the current assumption
  * Model is linear

When using the linear regression technique, the ideal model is always the one that has the smallest RSS value.  This is not necessarily true for all modeling techniques.  There are models where the minimum RSS is not the best model, however, if a model is linear then the model with the smallest RSS will always be the best model.  Linear regression models satisfy this condition



### Regression coefficients - Minimizing the RSS - Cont'd

````python
plt.figure(figsize=(16,8))


models = [(linreg.intercept, linreg.slope), (-2,0.2), (20,0), (20,-0.1)]

x_axis = np.arange(-10,320)

for (m, model) in enumerate(models):
    plt.subplot(2, 2, m+1)
    y_line = model[0] + model[1] * x_axis
    plt.scatter(advertising["TV"], advertising["sales"], alpha=0.2)
    RSS = 0 
    for i in range(advertising.shape[0]):
        y_hat = model[0] + model[1]  * advertising.iloc[i]["TV"]
        #plt.scatter(advertising.iloc[i]["TV"], advertising.iloc[i]["sales"], color='red', s=100)
        plt.vlines(advertising.iloc[i]["TV"], advertising.iloc[i]["sales"], y_hat, color='red')    
        RSS += (y_hat - advertising.iloc[i]["sales"])**2
    plt.plot(x_axis, y_line)
    plt.title( "RSS={}".format(round(RSS,1)), fontsize=18)
    
plt.legend()
        
````
