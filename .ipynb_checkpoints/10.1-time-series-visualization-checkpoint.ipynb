# 10.1  Time Series Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
import scipy.stats as stats
%matplotlib inline
```

### Visualizing Time Series

Using a graph to visualize time series data can make certains points of the data more apparent. When looking at a time series graph, it helps display the patterns (trend and seasonality) of the data, any outliers from the pattern, and changes in the data over time.

![Time Series Graphs](http:///128.171.10.138:4000/assets/ipynb/week_10/images/types_plot.png)

We can generate or load a dataframe using a date range and some values for an example graph.

```python
df = pd.DataFrame({'value': [5, 12, 15, 23, 25, 31, 35, 40, 38, 50, 55, 61, 10, 20, 16, 19, 31, 26, 35, 41, 39, 56, 50, 66, 
          15, 18, 15, 22, 25, 35, 33, 45, 42, 61, 57, 72, 22, 25, 17, 23, 28, 37, 50, 46, 56, 74, 66, 80]})
date_range = pd.date_range(start ="2004-01", end="2007-12", freq="MS")
len(date_range)
df.index = date_range
df.head()
```

```python
plt.figure(figsize=(20,8))
plt.plot(df)
```

In the above graph, there are values for four different years from the beginning of 2004 to the end of 2007. We can split the graph into four smaller graphs that highlight the values of each of the different years. By splitting the graphs up, we can see possible trends or seasonalities for a subset of the years.

```python
plt.figure(figsize=(20,6))

plt.subplot(2,2,1)
plt.plot(df["2004"])
plt.subplot(2,2,2)
plt.plot(df["2005"])
plt.subplot(2,2,3)
plt.plot(df["2006"])
plt.subplot(2,2,4)
plt.plot(df["2007"])
```

```python
years = sorted(set(df.index.year))
years
```

Another method of visualizing time series is to plot each of the years onto one graph. By plotting each year, it can make it much easier to identify trends or seasonalities when each plot of a year lies right next to the others. The graphs shows that there is a general upwards trend each year.

```python
plt.figure(figsize=(20,6))
orderedMonths = ["Jan", "Feb", "Mar", 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

    
for year in years:
    plt.plot(df[str(year)].reset_index()["value"])

plt.yticks([],[])
_ = plt.xticks(range(12), orderedMonths)
```

We can also add months and years to the dataframe to visualize the time series in a different way.

```python
df['month'] = df.index.strftime('%b')
df['year'] = df.index.year
df.head()
```

The dataframe can be transformed into a pivot table as another form of visualizing a time series.

```python
df_piv = df.pivot(index='year', columns='month', values='value')
df_piv.head()
```

By ordering the months, we can see the specific values for each year and each month. We can now see the trends and seasonalities based on how the values change over the year. Pivot tables also shows how much the value changes specifically instead of a general increase or decrease from the graphs.

```python
orderedMonths = ["Jan", "Feb", "Mar", 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_piv = df_piv.reindex(columns=orderedMonths)
df_piv.head()
```

### Polar Plot

- Polar plot is the plot of data using theta $\theta$ and radius $r$ instead of $x$ as a function of $y$.
- Polar plot in Time Series Visualization is very useful in viewing seasonal trends in data. Being able to view the data in this way is easier to notice seasonality. Especially when the plot is used to sample individual months over time. The focality or radius represents the value at that point in time.

```python
times = pd.date_range("01/01/2016", "12/31/2016")
rand_nums = np.random.rand(len(times),1)
df = pd.DataFrame(index=times, data=rand_nums, columns=['value'])

ax = plt.subplot(projection='polar')

t = mdates.date2num(df.index.to_pydatetime())
y = df['value']
tnorm = (t-t.min())/(t.max()-t.min())*2.*np.pi
ax.plot(tnorm, y, linewidth=0.8)

orderedMonths = ["Jan", "Feb", "Mar", 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
angles = np.linspace(0, 2*np.pi, 12, endpoint=False)
angles =  np.concatenate([angles, [angles[0]]])
ax.set_thetagrids(angles * 180/np.pi, orderedMonths)

plt.show()
```

If you examine the figure above, you can see clearly that there is neither seasonality nor trend. The values do not follow any type of pattern and appear random.

```python
plt.figure(figsize=(40, 6))
ref = np.arange(12)

orderedMonths = ["Jan", "Feb", "Mar", 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
angles = np.linspace(0, 2*np.pi, 12, endpoint=False)
angles =  np.concatenate([angles, [angles[0]]])

ax = plt.subplot(1, 1, 1, projection='polar')
ax.set_thetagrids(angles * 180/np.pi, orderedMonths)

    
for i in range(10):
    y = np.cos(ref) + i
    vals = np.concatenate([y, [y[0]]])
    plt.plot(angles, vals, label=str(i))
    
plt.legend(loc="right", bbox_to_anchor=(1.3, 0.55))
```

The plot above clearly shows that there is seasonality, there are recurring trends for each line and shows a pattern that can be predicted.

```python
orderedMonths = ["Jan", "Feb", "Mar", 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
theta = np.arange(0,361,10)
theta = theta/180.*np.pi
r = np.linspace(0.,0.8,len(theta))
ax = plt.subplot(111,projection = "polar")

ax.plot(theta,r)
angles = np.linspace(0, 2*np.pi, 12, endpoint=False)
ax.set_thetagrids(angles * 180/np.pi, orderedMonths)
plt.show()
```

The plot above shows that there is an upward trend throughout the year and indicates a very predictable pattern.

### Correlation Visualization
- Two variables are correlated if knowledge of one helps better predict the other
- scores range from -1 (negatively correlated) to 1 (positively correlated) through 0 (no correlation)
- popular coefficients used to measure correlation include the Spearman and Pearson coefficients

#### Pearson Correlation
The Pearson correlation test is parametric, so it makes certain assumptions about the data. In order for the results of a Pearson correlation test to be valid, the data must meet these assumptions:
- The sample is independently and randomly draw
- A linear relationship between the two variables is present
- When plotted, the lines form a line and is not curved
- There is homogeneity of variance
- The variables being used in the correlation test should be continuous and measured either on a ratio or interval sale, each variable must have equal number of non-missing observations, and there should be no outliers present.

```python
a = np.array([0, 0, 0, 1, 1, 1, 1])
b = np.arange(7)
stats.pearsonr(a, b)
```

The following result outputs the Pearson correlation coefficient followed by the p-value for testing non-correlation. The p-value roughly indicates the probability of an uncorrelated system producing datasets that have a Pearson correlation at least as extreme as the one computed from these datasets.

#### Spearman Correlation
Unlike the Pearson correlation test, the Spearman rank correlation is a non-parametric test that does not make any assumptions about the distribution of the data. The assumption for the Spearman rank correlation test is:
- There is a monotonic relationship between the variables being tested
- A monotonic relationship exists when one variable increases so does the other
- For the Spearman rank correlation, the data can be used on ranked data, if the data is not normally distributed, and even if the there is not homogeneity of variance.

```python
stats.spearmanr(a,b)
```

### Autocorrelation
We can calculate the correlation for time series observations with observations with previous time steps, called lags. Because the correlation of the time series observations is calculated with values of the same series at previous times, this is called a serial correlation, or an autocorrelation.

A plot of the autocorrelation of a time series by lag is called the Auto Correlation Function (ACF). This plot is sometimes called a correlogram or an autocorrelation plot. In the example below, we will run some autocorrelation visualization functions on air passenger data.

```python
import statsmodels.api as sm

data = pd.read_csv('AirPassengers.csv')
data.head()
```

```python
#visualizing passenger data over time

passenger_data = data['#Passengers']
plt.plot(passenger_data)
plt.show()
```

```python
from pandas.plotting import lag_plot

# a lag plot is a quick way to check for correlation

lag_plot(passenger_data)
plt.show()
```

Running a lag_plot shows the passenger data (t) on the x-axis against the passengers on the previous day (t-1) on the y-axis. We can see a cluster of observations along a diagonal line of the plot. It clearly shows a relationship or some correlation.

This process could be repeated for any other lagged observation, such as if we wanted to review the relationship with the last 7 days or with the same day last month or last year. This lag_plot is confirmed by the Pearson correlation test results shown below:

```python
from pandas import concat

corr_test = concat([passenger_data.shift(1), passenger_data], axis=1)
corr_test.columns = ['t-1', 't+1']
result = corr_test.corr()
print(result)
```

#### Autocorrelation Plot
We can plot the correlation coefficient for each lag variable.

This can very quickly give an idea of which lag variables may be good candidates for use in a predictive model and how the relationship between the observation and its historic values changes over time.

We could manually calculate the correlation values for each lag variable and plot the result, but in this case we will use statsmodels' acf model.

```python
# autocorrelation
sm.graphics.tsa.plot_acf(passenger_data, lags=40)
plt.show()
```

#### Partial Autocorrelation Plot
A partial autocorrelation is a summary of the relationship between an observation in a time series with observations at prior time steps with the relationships of intervening observations removed.

The autocorrelation for an observation and an observation at a prior time step is comprised of both the direct correlation and indirect correlations. These indirect correlations are a linear function of the correlation of the observation, with observations at intervening time steps.

It is these indirect correlations that the partial autocorrelation function seeks to remove. Without going into the math, this is the intuition for the partial autocorrelation.

The example below calculates and plots a partial autocorrelation function for the first 30 lags in the Air Passenger dataset using the plot_pacf() from the statsmodels library.

```python
sm.graphics.tsa.plot_pacf(passenger_data, lags=30)
plt.show()
```
