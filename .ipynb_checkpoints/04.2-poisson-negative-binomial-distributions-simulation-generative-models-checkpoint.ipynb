### The Poisson Distribution
    
The **Poisson distribution** is the **discrete probability distribution** of the number of events occurring in a fixed space or time period, given an average rate. **Discrete** data is data that can only take on particular values in a range (as opposed to a continuum). A **probability distribution** is an assignment of probabilities to the values of the random variable. 

The Poisson Distribution is used to answer questions like:

**"What is the probability that there will be 120 parking tickets handed out at UH in a given week?"**


By using the Poisson Distribution you can model the number of events in a fixed period of time or space. 

- The number of mushrooms on a Large Pizza 
- The number of noise complaints in Hale Aloha every week
- The number of NFL touchdowns scored every day



#### Properties of the Poisson Distribution

In the normal distribution we discussed earlier the mean and the variance were two separate parameters. The mean had no effect on the variance. With the Poisson Distribution there is a single parameter, $\lambda$, that determines the shape of the distribution and the location of the distribution on a graph.

The probability mass function ($pmf$) of the Poisson is given below: 

$$
p(X=x) = \frac{e^{_\lambda} \lambda^x}{x!}
$$



$X$ is Poisson if $X$ is the number of events in a given time period or space area and the events are **independent** $X$ such that the occurrence of an event has no effect on the likelihood of the following events. With the Poisson is $X$ is **countably infinite** {0,1,2,3...} and non-negative. In a countably infinite set you can count off each of the elements and the counting will go on infinitely, however there will be a finite amount of time to reach a particular element in the set. 
<img src= 'images/poisson.png' style= 'width:500px; float:right;'>


In the Poisson **$\lambda$** is both the **mean** number of occurrences in the interval and the **variance**. $e$ is [Euler's constant](https://en.wikipedia.org/wiki/E_(mathematical_constant)), e $\approx$ 2.718, it is the limit of $(1 + 1/n)^n$ as n approaches infinity.


Unlike the Normal distribution the shape of the Poisson distribution is always changing and need not be symmetric. As $\lambda$ increases the Poisson distribution becomes more symmetrical, but for smaller $\lambda$ values there is a right skew. 

This graph shows the Poisson for different $\lambda$ values. Note how the shape and position of the distribution changes as $\lambda$ increases.

Notice how the distribution starts to look more normal as $\lambda$ gets large.

### Example

Lets look at an example of the Poisson in python using the numpy and scipy packages.

Suppose we want to model the number of clients that visit a store in a hour using a distribution, assume that the number of clients to visit a store per hour follows a Poisson distribution with $\lambda = 10$. This implies that, **each occurrence is independent** of the other occurrences, a client visit during an hour should have no effect on the next hour, the **mean number of clients visiting the store per hour is constant** over all hours, and the number of visitor per hour is **countably infinite**.  

The probability distribution of the random variable $X$ representing the number of clients in the store is:

````python
plt.figure(figsize=(10,5)) # Create a plot with size 12 x 7 
x = np.arange(0, 30, 1) # Generate a list of values up to 30 (non inclusive) starting with 0 with increments of 1
pmfs_x = poisson.pmf(x, 10) # For each value in x generate give a P(X) using a Poisson with a mean of 10 
plt.plot(x, pmfs_x,) # Plot x against the pmfs_x values generated by the Poisson, in other words graph the Poisson
_ = plt.xlabel("$X$") # Assign X axis Label
_ = plt.ylabel("$P(X)$") # Assign y axis Label
````

Below in orange is a same plot as above with an additional Poisson Distribution with $\lambda = 20$, modeling the same scenario above but where the **average number of clients** that visit a store in a hour is **20**. Note how the **position** of the distribution has changed. You can also see the distribution become **more symmetric** since a larger $\lambda$ was used to generate it. 

````python
plt.figure(figsize=(10,5)) # Create a plot with size 12 x 7 
x = np.arange(0, 30, 1) # Generate a list of values up to 30 (non inclusive) starting with 0 with increments of 1
pmfs_x = poisson.pmf(x, 10) # For each value in x generate give a P(X) using a Poisson with a mean of 10 
plt.plot(x, pmfs_x,) # Plot x against the pmfs_x values generated by the Poisson, in other words graph the Poisson
x = np.arange(0, 30, 1) # Generate a list of values up to 31 (non inclusive) starting with 0 with increments of 1
pmfs_x = poisson.pmf(x, 20) # For each value in x generate give a P(X) using a Poisson with a mean of 20 
plt.plot(x, pmfs_x,) # Plot x against the pmfs_x values generated by the Poisson, in other words graph the Poisson
_ = plt.xlabel("$X$") # Assign X axis Label
_ = plt.ylabel("$P(X)$") # Assign y axis Label
````

### Link between the mean and the variance

As we have seen, in the **normal** distribution the **mean** and **variance** are **separate** parameters. Each has no effect on the other. However, with the **Poisson** the **mean** and **variance** are both modeled with the **same** parameter $\lambda$.

Below are the distributions of the Poisson highlighting how the **variance** changes with the **mean**. 

````python

plt.figure(figsize=(15,5))

x = np.arange(0, 23, 1)
pmfs_x = poisson.pmf(x, 10)
plt.plot(x, pmfs_x)


x = np.arange(50, 150, 1)
pmfs_x = poisson.pmf(x, 100)
plt.plot(x, pmfs_x)


x = np.arange(100, 300, 1)
pmfs_x = poisson.pmf(x, 200)
plt.plot(x, pmfs_x)

x = np.arange(150, 450, 1)
pmfs_x = poisson.pmf(x, 300)
plt.plot(x, pmfs_x)


x = np.arange(200, 600, 1)
pmfs_x = poisson.pmf(x, 400)
plt.plot(x, pmfs_x)
````

### Notes about the Poisson

All data can not be modeled using the Poisson distribution. When count data has **wide dispersion** the variance may be very different from the mean, and the **Poisson can not** be used. 

Though this seems like a severe limitation of the distribution, it can none-the-less be applied in a variety of different scenarios. Telecommunications companies use the Poisson to mode the number of phone calls they may receive during a particular period. DNA researchers use the Poisson to model the number of mutations in the human genome.



For values of $\lambda > 40$, the Poisson can approximate the Gaussian with $\mu= \lambda$ and $\sigma = \sqrt{\lambda}$. You can [show](https://www.roe.ac.uk/japwww/teaching/astrostats/astrostats2012_part2.pdf) how this works mathematically by proving the Gaussian distribution is a limit of the Poisson distribution. Working with the Poisson can be simpler than working the Gaussian because there is only a single parameter, $\lambda$. 


The Graph below compares a Gaussian compared to a Poisson distribution with the parameters described above. Note how closely the match one another. 

````python
import matplotlib.mlab as mlab
import math
plt.figure(figsize=(12,5))
mu = 40
variance = 40
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x,mlab.normpdf(x, mu, sigma))
x = np.arange(10,60,1); 
pmf_x = poisson.pmf(x,40); 
plt.plot(x,pmf_x)
````

### The Negative Binomial Distribution

- The _negative_ _binomial_ ( $NB$ ) is a discrete distribution of the number of trials with binary trials (success and failure, true and false, healthy and diseased, etc.) required before the $k^{th}$ success.

The _negative_ _binomial_  ( $NB$ ) is a discrete probability distribution of the number of successes in a sequence of independent and identically distributed binary trials (Bernoulli trials) before a specified number of failures occurs. Or another less common way to put it the _negative_ _binomial_  a discrete distribution of the number of trials with binary trials (success and failure, true and false, healthy and diseased, etc.) required before the $k^{th}$ success.

For example, the $NB$ can be used model the number of times you need to roll a die to get five ones,to model the number of ads you need you need to display on a website to have 100 clicks, or to model the number of the lottery tickets you need to buy before winning the jackpot. The _negative_ _binomial_ can be used applied to scenarios where the number that is varied is the number of trials and the fixed number is the number of successes. 

It is key to remember the difference between the Negative Binomial and the Binomial. With the Binomial the random variable is the number of successes in a fixed number of trials. In the Negative Binomial the random variable is the number of trials with a fixed number of successes.

As mentioned above, you can also think of the negative binomial as the total number of failures before you see $k$ successes, or the number of trials before a success. Both operate on the same principle but the equations required will be slightly different as you either vary the number of failures (total number of trials - successes), or the number of trials (successes + failures).


  - This definition leads to slightly different equations, but the intuition is similar. 

### The Negative Binomial pmf

$$
P(X=k) =  {{n-1}\choose{k-1}}p^k (1-p)^{n-k}
$$

This equation is for modeling by varying trials (total number). 

- $p$ probability of success for a trial
- $k$ number of successes
- $X$ the random variable that is the total number of trials before $k$ successes

#### Properties of the Negative Binomial Distribution

For the _negative_ _binomial_ to apply to distribution a few criteria must be met. The sequence of trials will independent, that is the order in which the trials are conducted should not matter. Each of the trials should have binary results, success or a failure. Binary trials of this type are called Bernoulli trials, a random experiment with exactly two possible outcomes. The probability of success should not change from trial to trial, it should remain a constant $p$. $X$, the total number of trials before $k$ successes, can take any value from 0 to inifinity [0,1,2,3...]. The values of $X$ is therefore countably infinite. And finally the number of trials is not set as in the Binomial.

Suppose you want to determine the number of times you must roll a dice for it to land on 1, 7 times. This can be modeled using the Negative Binomial as the probability, p, of landing on 1 is consistently (1/6), and the number of success is fixed,k, is 7.

Below is the python code for modeling such a distribution. 
   - p = 1/6
   - X can be any value such such as $x \in [7, \infty]$


````python
from scipy.stats import nbinom
plt.figure(figsize=(8,5))
x = np.arange(0, 120, 1)

pmfs_x = [nbinom.pmf(val, 7, 1/6)  for val in x]

plt.plot(x, pmfs_x)

````

Below is the python code for modeling the same problem but with only 3 rolls of 1 needed. 
   - p = 1/6
   - X can be any value such such as $x \in [3, \infty]$

````python
from scipy.stats import nbinom
plt.figure(figsize=(8,5))
x = np.arange(0, 120, 1)

pmfs_x = [nbinom.pmf(val, 3, 1/6)  for val in x]

plt.plot(x, pmfs_x)

````

### Expected Value and Value with Highest Probability

With a symmetrical distribution, if we were to the same calculation as above, the probability on the left and right would be identical. Values occur in the same frequency on both sides of the mean with a symmetrical distribution. With a symmetrical distribution the expected value becomes the most probable values in the long run. 

With an asymmetrical distribution, values on one side of the most probable value occur with more probability than on the other side, as seen above. This means that the values with the highest probability is not necessarily the expected value. 

In the case above with 7 successes, we have
```python
a = sum([nbinom.pmf(x,  7, 1/6) for x in range(0,30)])
b = sum([nbinom.pmf(x,  7, 1/6) for x in range(30,150)])

The total probab of value on the left is 0.3932519887746541
The total probab of value on the right is 0.606747372808539
The total sum of the probabilities undr the curve is 0.9999999995481055
```

Therefore more values will be generate by values on the right side of the most likely value.  This means the expected value will be on the right.

# Introduction to Simulation

In this section, we are going to explore random or stochastic models.  Probabilistic modeling relies on aspects of human cognition.
We will also cover graphical models which are very intuitive and often require less theory to implement.  Graphical models use  generative approach to modeling meaning that we reason about the problem by studying how data can be generated.

### Example: Constructing a Generative Model

#### The Situation
Consider the situation where a patient goes into the doctor with symptoms of an illness.  For simplicity, let's assume that there are only three possible illnesses: bronchitis, common cold, and seasonal allergies.  Also for simplicity, let us only consider three possible symptoms: fever, coughing, and sneezing.

#### Assumptions About the Generative Model

We often know how diseases are related to the symptoms.  For example, fever is more likely in common cold than allergies.

Let us assume the folloing is true for having a cold: 
  * 10% of patients with a cold have a fever 
  * 50% of the patients with a cold sneeze 
  * 40% of the patients with a cold cough

We want to use this information to find the probability that a patient has a certain illness given their symptoms, but first, we are going to create a generative model for our situation.

### Generative models Representation
---

A *generative model* uses a hierarchical construct, which provides a recipe for generating data.  It is commonly represented using a graph structure (tree).  In it's simplest form, it shows random variables and their dependencies.  Random variables are represented as vertices in the graph and dependencies are encoded as edges.

### Generative Model for the Common Cold
<img src="http://128.171.10.138:4000/assets/ipynb/week_4/images/gen_model_cold.jpg" style="width:500px; float:right;">

Let us consider 10 people with a cold.  The model to the right shows us that 1 has a fever, 4 have a cough, and 5 people are sneezing. For simplification, the symptoms are non-overlapping.

Arrows (edges) illustrate the dependencies between the disease(cold) and its symptoms.

Using this model, we can now generate new data.  The code below uses numpy to randomly select a symptom for a patient who has a cold, using the proportions stated above (fever 0.1, cough 0.4, sneeze 0.5).  This is a simulation.  If we do this many times, we can simulate many patients who have a cold and record all of their symptoms.

````python
# generative model for cold example
import numpy as np

p_fever = 0.1   # number of people with a cold & fever
p_cough = 0.4   # number of people with a cold & a cough
p_sneeze = 0.5   # number of people with a cold & a sneeze

# or
cold = ['fever', 'coughing', 'sneezing']

patient_symptom = np.random.choice(cold, p = [p_fever, p_cough, p_sneeze])

print("We generate a patient who has cold and is:   %s" % (patient_symptom))
````

````python
disease = ['Bronchitis', 'Cold', 'Allergies']
disease_p = [0.1,0.4,0.5]

patient_disease = np.random.choice(disease, p= disease_p)

print("We generate a patient who has:   %s" % (patient_disease))
````

#### Example Generative Model
The following code generates a data frame of 1000 randomly selected illnesses and symptoms.  Each row represents a patient with a randomly generated illness and symptom.

````python
probs_symptoms = {"bronchitis": {"fever": 0.2, "cough": 0.8}, 
         "cold": {"fever": 0.1, "cough": 0.4, "sneeze": 0.5}, 
         "allergies": {"fever": 0.1, "cough": 0.4, "sneeze": 0.5}
        }

prob_illnesses = {"bronchitis": 0.1, "cold": 0.4, "allergies": 0.5}

illness_samples = np.random.choice(list(prob_illnesses.keys()), p= list(prob_illnesses.values()), size= 1000)
illness_df = pd.DataFrame({"illnesses": illness_samples})

def sample_symptom(illness):
    return np.random.choice(list(probs_symptoms[illness].keys()), p = list(probs_symptoms[illness].values()))
    
sample_symptom("cold")
illness_df.applymap(sample_symptom)
illness_df["symptom"] = illness_df["illnesses"].apply(sample_symptom)
illness_df.head()
````

### Another Example: Simulating Sales Revenue

Imagine you have a website that sells a product and according to your tracking, the distribution of visitors to your site appears to be normally distributed with an average of 2000 unique visitors a week and a standard deviation of 157.

You also know that each visitor has a probability of 0.05 of buying from your site and that clients who purchase on your site tend to spend, on average, \$140, with a standard deviation of 25.  The distribution of sales per client also seems normal.

We can use a simulation to see what do your sales at the end of the year (52 weeks) look like.


### Step 1: Simulating Unique Visitors

Since the distribution of visitors to your site looks normal, we can sample the number of visitors using the given mean and standard deviation.

Using `numpy.random` we can simulate the number of visitors each week `np.random.normal(2000, 157)` and put it in a dataframe under the column `nb_clients`


````python
#import pandas as pd
sim_data = pd.DataFrame({"Week": range(1,53)})

sim_data["nb_clients"] = np.random.normal(2000, 157, size=sim_data.shape[0])
sim_data["nb_clients"] = sim_data["nb_clients"].apply(int)
sim_data.head()
````

### Step 2: Simulating the number of visitors who buy

We can simulate this data best by using binomial distribution since our options are 
* success: visitor buys ($p=0.05$)
* failure: visitor doesn't buy ($p=0.95$)

Each week we have x visitors, which are our trials ($n$)

If we use `np.random.binomial(nb_clients, 0.05)` on nb_clients from each row, it will generate the number of buyers for each week.

````python
def get_number_purchases(x, prob=0.05):
    return np.random.binomial(x, prob)

sim_data["nb_buyers"] = sim_data["nb_clients"].apply(get_number_purchases)
sim_data.head()
````

### Step 3: Sampling total weekly sales

To sample total weekly sales, we can use a normal distribution again as these values are based on magnitude.  For our example, the clients who purchase on our site tend to spend on average \\$140 with a standard deviation of \\$25.  So we can generate as many transactions as there are buyers in the `nb_buyers` column.  

First a random sample is drawn from a normal with mean ($\mu = 140$ and $\sigma = 25$) for each of the buyers.
`np.random.normal(140, 25, nb_buyers)`

Then we can sum up the sales for that week:
```python
def total_weekly_sales(nb_buyers, mean = 140, std=25):
    return sum(np.random.normal(140, 25, nb_buyers))
```
We apply this to all the `nb_buyers` rows to get the `total_sales` column.

````python
def total_weekly_sales(nb_buyers, mean = 140, std=25):
    return sum(np.random.normal(140, 25, nb_buyers))
    
sim_data["total_sales"] = sim_data["nb_buyers"].apply(total_weekly_sales)
sim_data.head()
````

### Step 4: Inferring total sales

To get the total sales estimate for the year, we simply sum the simulated weekly sales as shown below.

````python
sim_data["total_sales"].sum()
````

````python
plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)
plt.hist(sim_data["nb_clients"])
plt.xlabel("NB Clients", fontsize=18)

plt.subplot(1, 3, 2)
plt.hist(sim_data["nb_buyers"])
plt.xlabel("NB Purchases", fontsize=18)

plt.subplot(1, 3, 3)
plt.hist(sim_data["total_sales"])
plt.xlabel("Total Sales", fontsize=18)
````

### Follow-up Question

Is this a sufficient *scenario* to get a good estimate for our yearly sales?
Think about it. You have one year of "data," but how do you determine whether you just had a hot year or a really bad one without other years to compare to?  

So we need to look at **many** years of data.

When we are simulating any sort of data, we need more than one sample set to see a trend. This is because we are interested in analyzing the long-term frequency of our data. 

So next, we will repeat the earlier example with 100 iterations instead of 1, and, for our convenience, add a new column to keep track of iterations.

### Simulating Sales
Applying earlier functions to populate our new dataset:

````python
sim_data_iteration = pd.DataFrame({ "Iteration": ["1"] * 52, "Week": range(1,53)})
sim_data_iteration["nb_clients"] = pd.Series(np.random.normal(2000, 157, size=sim_data.shape[0])).apply(int)
sim_data_iteration["nb_buyers"] = sim_data_iteration["nb_clients"].apply(get_number_purchases)
sim_data_iteration["total_sales"] = sim_data_iteration["nb_buyers"].apply(total_weekly_sales)

total_data = sim_data_iteration
total_data.head()
````

Now repeating that step 99 more times...

````python
for i in range(2, 101):
    sim_data_iteration = pd.DataFrame({ "Iteration": [str(i)] * 52, "Week": range(1,53)})
    sim_data_iteration["nb_clients"] = pd.Series(np.random.normal(2000, 157, size=sim_data.shape[0])).apply(int)
    sim_data_iteration["nb_buyers"] = sim_data_iteration["nb_clients"].apply(get_number_purchases)
    sim_data_iteration["total_sales"] = sim_data_iteration["nb_buyers"].apply(total_weekly_sales)
    total_data  = total_data.append(sim_data_iteration)

total_data.shape
````

### Analyzing our larger dataset

Now that we have a dataset of a much greater scale, it would be hard to display all the rows. This is where we can utilize DataFrame methods to help us verify that our data was successfully simulated.

````python
total_data.groupby("Iteration", sort=False).count().iloc[:5]
````

Using groupby and head methods to sort, and the iloc attribute to reduce the output size, we can take a look at a subset of our data.

````python
total_data.groupby("Iteration").head(3).iloc[:9]
````

Now that we have generated a sizable dataset to work with and examined it from different angles, we can plot it again.

The histograms produced now are more consistent and give us a better idea of what our data is telling us.

````python
plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)
plt.hist(total_data["nb_clients"])
plt.xlabel("NB Clients", fontsize=18)

plt.subplot(1, 3, 2)
plt.hist(total_data["nb_buyers"])
plt.xlabel("NB Purchases", fontsize=18)

plt.subplot(1, 3, 3)
plt.hist(total_data["total_sales"])
plt.xlabel("Total Sales", fontsize=18)
````

### More test cases

Our charts look much better now and have a shape of what a distribution should look like. However, with some ambiguity in the peak frequencies, we might consider doing more simulations on more data.

````python
for i in range(101, 1001):
    sim_data_iteration = pd.DataFrame({ "Iteration": [str(i)] * 52, "Week": range(1,53)})
    sim_data_iteration["nb_clients"] = pd.Series(np.random.normal(2000, 157, size=sim_data.shape[0])).apply(int)
    sim_data_iteration["nb_buyers"] = sim_data_iteration["nb_clients"].apply(get_number_purchases)
    sim_data_iteration["total_sales"] = sim_data_iteration["nb_buyers"].apply(total_weekly_sales)
    total_data  = total_data.append(sim_data_iteration)

total_data.shape
````

### Question:

What if we wanted to see the growth of our total sales in terms of time? 

We can use the cumsum method to see the cumulative total sales over time (weekly).

````python
total_data["cumulative_sum"] = total_data[["Iteration", "total_sales"]].groupby("Iteration").cumsum()
total_data[total_data["Iteration"] == "1" ].head()
````
